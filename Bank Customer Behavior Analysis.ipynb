```python
# Extraction the Distribution of unique catgories in categorical colums

categorical_col = df[['BRId' , 'GenderId' , 'IAId', 'Amount of Credit Cards' , 'Nationality' , 'Occupation' ,'Fee Structure', 'Loyalty Classification', 'Properties Owned' , 'Risk Weighting' ,  'Income Band']].columns
```

```python
for col in categorical_col:
  print(f"value counts for '{col}' :")
  display(df[col].value_counts())
```

```python
for i , predictor in enumerate(categorical_col):
    plt.figure(i, figsize=(8, 5))
    sns.countplot(data = df , x = predictor , color = "Skyblue")
    plt.title(f"Count Plot of {predictor}", fontsize=14)
    plt.xticks(rotation=45)  # Rotate x-labels for better readability
    plt.xlabel(predictor)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()
```

```python
# Histogram of value counts for different occupaions

for col in categorical_col:
  if col  == 'Occupation':
    continue
  plt.figure(figsize = (8,4))
  sns.histplot(df[col])
  plt.title('Histogram of Occupation Count')
  plt.xlabel(col)
  plt.ylabel("count")
  plt.show()
```

Numerical Analysis


```python
numerical_col  = ['Estimated Income' , 'Superannuation Savings' , 'Credit Card Balance', 'Bank Loans' ,'Checking Accounts' , 'Saving Accounts' , 'Foreign Currency Account' ,'Business Lending' ,'Properties Owned' ]
```

```python
# Univariant Analysis and Visualisation

n_cols = 3  # Number of columns for subplots
n_rows = (len(numerical_col) + n_cols - 1) // n_cols  # Calculate number of rows needed

plt.figure(figsize=(15, n_rows * 5))  # Adjust figure size based on number of rows

for i, col in enumerate(numerical_col):
  plt.subplot(n_rows, n_cols, i + 1)
  sns.histplot(df[col], kde=True)
  plt.title(col)

plt.tight_layout()
plt.show()
plt.close()
```

```python
# Heatmap

corelation_matrix = df[numerical_col].corr()
plt.figure(figsize = (8,6))
sns.heatmap(corelation_matrix, annot = True, cmap = 'crest' , fmt = "2f")
plt.title('corelation_metrics')
plt.show()
```
